```python
# Discord Unneeded Data
WMT_close = pd.DataFrame(WMT.Close)
TGT_close = pd.DataFrame(TGT.Close)
AMZN_close = pd.DataFrame(AMZN.Close)
BBY_close = pd.DataFrame(BBY.Close)
COST_close = pd.DataFrame(COST.Close)
GM_close = pd.DataFrame(GM.Close)
```

```python
# Compute Moving Averages 
WMT_close['MA_WMT_30'] = WMT_close.Close.rolling(30).mean()
TGT_close['MA_TGT_30'] = TGT_close.Close.rolling(30).mean()
AMZN_close['MA_AMZN_30'] = AMZN_close.Close.rolling(30).mean()
BBY_close['MA_BBY_30'] = BBY_close.Close.rolling(30).mean()
COST_close['MA_COST_30'] = COST_close.Close.rolling(30).mean()
GM_close['MA_GM_30'] = GM_close.Close.rolling(30).mean()

#With Lag
# GM_close['MA_GM_30'] = GM_close.Close.rolling(30, center=True).mean()
```
